```python
import pandas as pd

df = pd.read_csv("house_price.csv")
df.head()
```

```python
print(df.shape)
print(df.columns)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = df["area"].values
y = df["price"].values
```

```python
X = (X - np.mean(X)) / np.std(X)
y = (y - np.mean(y)) / np.std(y)
```

```python
m = 0.0
b = 0.0

learning_rate = 0.01
epochs = 1000

n = len(X)
losses = []
```

```python
for epoch in range(epochs):
    
    y_pred = m * X + b

    loss = np.mean((y - y_pred) ** 2)
    losses.append(loss)

    dm = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    m -= learning_rate * dm
    b -= learning_rate * db
```

```python
print(f"Final learned equation: y = {m:.4f}x + {b:.4f}")
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss vs Epoch")
plt.show()
```
